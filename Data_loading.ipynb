```python
import pandas as pd
import json
```

```python
df_banner = pd.read_excel('data/Копия banner.xlsx')
df_banner
```

```python
df_native = pd.read_csv('data/Копия native.csv', usecols=[1,2,3,4,5])
df_native
```

```python
df_organic = pd.read_csv('data/Копия organic_traffic.csv', usecols=[1,2,3,4,5], header = None, 
                         names = ['user_id', 'funnel_stage', 'timestamp', 'profit', 'os'])
df_organic
```

```python
df_other = pd.read_csv('data/Копия other.csv', usecols=[1,2,3,4,5])
df_other
```

```python
df_partner = pd.read_csv('data/Копия partner_network.csv', usecols=[1,2,3,4,5], header = None, 
                         names = ['user_id', 'funnel_stage', 'timestamp', 'profit', 'os'])
df_partner
```

```python
df_rich = pd.read_excel('data/Копия rich.xlsx')
df_rich
```

```python
df_search = pd.read_csv('data/Копия search.csv', usecols=[1,2,3,4,5])
df_search
```

```python
df_video = pd.read_csv('data/Копия video.csv', usecols=[1,2,3,4,5])
df_video
```

```python
with open('data/Копия advertising_information.json', 'r') as file:
    adv_inf = json.load(file)
print(adv_inf)
```

```python
df = pd.concat([df_banner, df_native, df_organic, df_other, df_partner, df_rich, df_search, df_video], 
               axis = 0, keys=['banner', 'native', 'organic_traffic', 'other', 'partner_network', 'rich',
                                                 'search', 'video']).reset_index(level=0).rename(columns={"level_0": "ad_channel"})
df
```

```python
df_adv_info = pd.DataFrame(adv_inf)
df_adv_info
```

```python
df_result = df.merge(df_adv_info, on="ad_channel", how="left")
df_result
```

```python
new_columns_order = ['user_id', 'ad_channel', 'funnel_stage', 'price', 'profit', 'os', 'timestamp']

df_result = df_result.reindex(columns=new_columns_order)
df_result
```

```python
df_result = df_result.sort_values(by = 'timestamp', ascending = True)
df_result
```

```python
df_result.to_csv('data/advertising_campaign.csv')
```
